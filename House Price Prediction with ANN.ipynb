```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
df = pd.read_csv('D:/dataset/datahouse.csv')
```

```python
df.head()
```

```python
bedrooms = df['bedrooms']
bathrooms = df['bathrooms']
luas = df['sqft_living']
floor = df['floors']
prices = df['price']
```

```python
rumah_all = []

for i in range(len(bedrooms)):
    rumah = []
    rumah.append(bedrooms[i])
    rumah.append(bathrooms[i])
    rumah.append(luas[i])
    rumah.append(floor[i])
    
    rumah_all.append(rumah)
```

```python
print(rumah_all)
rumah_all = np.array(rumah_all)
prices = np.array(prices)
```

```python
model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(units = 1, input_shape = [4])
])
```

```python
model.compile(
      loss='mae',
      optimizer='sgd',
      metrics=['accuracy']
)
```

```python
model.fit(
    rumah_all, prices, epochs=100
)
```

```python
kamar_tidur = int(input("Masukan jumlah kamar tidur yang anda inginkan : "))
kamar_mandi = int(input("Berapa banyak kamar mandi yang anda inginkan : "))
luas_rumah = int(input("Anda mau rumah seluas apa? "))
jumlah_lantai = int(input("Anda ingin rumah berapa lantai? "))

prediksi_harga = model.predict([[kamar_tidur, kamar_mandi, luas_rumah, jumlah_lantai]])
print("Rumah impian anda kira2 harganya {} USD".format(prediksi_harga[0][0]))
```
